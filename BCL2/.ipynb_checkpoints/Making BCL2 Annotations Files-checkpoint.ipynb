```python
import pandas as pd
from poola import core as pool
import seaborn as sns
import gpplot
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
beagleAG = pd.read_csv("crisprbe-guides_bcl2_BEAGLE_A-G.txt",sep='\t')
beagleCT = pd.read_csv("crisprbe-guides_bcl2_BEAGLE.txt",sep='\t')
bcl2_read_counts = pd.read_csv('counts-ALG_MOLM13_CP1589_CP1590.txt',sep='\t')
```

```python
bcl2_read_counts.head(10)
```

```python
#save controls separately because they aren't in the BEAGLE files
controls = bcl2_read_counts[~(bcl2_read_counts['Construct IDs'].str.contains('BCL2'))]
```

```python
controls.shape
```

```python
#edit controls df so that it can be concatenated with the annotation file
controls = controls[["Construct Barcode", 'Construct IDs']]
controls['Amino Acid Edits'] = controls['Construct IDs']
controls['Mutation Category'] = controls['Construct IDs']
controls = controls.rename(columns = {"Construct Barcode":"sgRNA Target Sequence",'Construct IDs': 'Target Gene Symbol'})
controls
```

```python
beagleAG.shape
```

```python
#filter for guides that were used in the library
beagleAG = beagleAG[(beagleAG['sgRNA Target Sequence'].isin(bcl2_read_counts['Construct Barcode']))]
beagleAG.shape
```

```python
#add controls back in
beagleAG = pd.concat([beagleAG, controls])
beagleAG.shape
```

```python
beagleCT.shape
```

```python
#filter for guides that were used in the library
beagleCT = beagleCT[(beagleCT['sgRNA Target Sequence'].isin(bcl2_read_counts['Construct Barcode']))]
#add controls back in
beagleCT = pd.concat([beagleCT, controls])
beagleCT.shape
```

```python
beagleAG.to_csv('BCL2_ABE_annotations.csv')
beagleCT.to_csv('BCL2_CBE_annotations.csv')
```
